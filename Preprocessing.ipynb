```python
import pandas as pd
import re
from contractions import fix
from textblob import TextBlob
import nltk
from nltk.corpus import stopwords
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

def expand_contractions(text):
    return fix(text)

def correct_spelling(text):
    return str(TextBlob(text).correct())

def remove_rare_words(words, threshold=1):
    word_counts = Counter(words)
    return [word for word in words if word_counts[word] > threshold]

def remove_common_words(words, common_words):
    return [word for word in words if word not in common_words]

stemmer = PorterStemmer()
```

```python
nltk.download('stopwords')
common_words_en = set(stopwords.words('english'))
```

```python
df["remove_punctuation"] = df["content"].apply(lambda x: remove_punctuation(x))
df["expand_contractions"] = df["content"].apply(lambda x: expand_contractions(x))
df["correct_spelling"] = df["content"].apply(lambda x: correct_spelling(x))

nltk.download('punkt_tab')
df["tokens"] = df["content"].apply(lambda x: nltk.word_tokenize(x))
df["porter_stem"] = df["tokens"].apply(lambda x: [stemmer.stem(word) for word in x])

df["remove_rare_words"] = df["porter_stem"].apply(lambda x: remove_rare_words(x))
df["remove_common_words"] = df["porter_stem"].apply(lambda x: remove_common_words(x, common_words_en))
```
