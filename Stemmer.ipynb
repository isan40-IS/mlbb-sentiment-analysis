```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
```

```python
df = pd.read_csv("data.csv")
print(df.head())
```

```python
# Initialize the Porter Stemmer
stemmer = PorterStemmer()

# Tokenize the sentence into words
nltk.download('punkt_tab')
df["tokens"] = df["content"].apply(lambda x: nltk.word_tokenize(x))

# Apply stemming to each word
df["porter_stem"] = df["tokens"].apply(lambda x: [stemmer.stem(word) for word in x])
df["porter_stem"]
```

```python
# Initialize the Lancaster Stemmer
lancaster_stemmer = LancasterStemmer()

# Apply Lancaster stemming to each word
df["lancaster_stem"] = df["tokens"].apply(lambda x: [lancaster_stemmer.stem(word) for word in x])
df["lancaster_stem"]
```
