### Import Library

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
from collections import Counter
import numpy as np
```

```python
# 1. Dataset dari slide Stanford
corpus = [
    "Ekonomi Indonesia pada kuartal I-2025 tumbuh 4,87% dan penopang signifikan pertumbuhan tersebut adalah ramadan. Temuan itu mendorong TikTok, Tokopedia dan TikTok Shop memperkuat sinergi guna memicu aktivitas ekonomi lokal terus melaju dengan memberikan pengalaman ekstra pada ramadan tahun ini.", 
    "Communications Lead, TikTok Indonesia, Edwin Lengkei menjelaskan TikTok, Tokopedia dan TikTok Shop menghadirkan kampanye Ramadan Ekstra Seru 2026 yang menghubungkan kreator, komunitas, penjual dan pembeli dalam sebuah ekosistem digital terintegrasi untuk memperoleh inspirasi, hiburan sampai pengalaman bertransaksi aman, nyaman dan berdampak nyata.",
    "Menurutnya ramadan selalu menjadi momen istimewa di TikTok. Tahun lalu, lebih dari 21,9 juta video diunggah selama ramadan, mencerminkan tingginya partisipasi komunitas TikTok di Indonesia."
]

# 2. Representasi Vektor (Hanya fokus pada kata 'ramadan' dan 'TikTok')
vectorizer = CountVectorizer(vocabulary=['ramadan', 'tiktok'])
X = vectorizer.fit_transform(corpus)
```

```python
# 3. Hitung Cosine Similarity
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['ramadan', 'tiktok'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

```python
documents = [
    "Ekonomi Indonesia pada kuartal I-2025 tumbuh 4,87% dan penopang signifikan pertumbuhan tersebut adalah ramadan. Temuan itu mendorong TikTok, Tokopedia dan TikTok Shop memperkuat sinergi guna memicu aktivitas ekonomi lokal terus melaju dengan memberikan pengalaman ekstra pada ramadan tahun ini.", 
    "Communications Lead, TikTok Indonesia, Edwin Lengkei menjelaskan TikTok, Tokopedia dan TikTok Shop menghadirkan kampanye Ramadan Ekstra Seru 2026 yang menghubungkan kreator, komunitas, penjual dan pembeli dalam sebuah ekosistem digital terintegrasi untuk memperoleh inspirasi, hiburan sampai pengalaman bertransaksi aman, nyaman dan berdampak nyata.",
    "Menurutnya ramadan selalu menjadi momen istimewa di TikTok. Tahun lalu, lebih dari 21,9 juta video diunggah selama ramadan, mencerminkan tingginya partisipasi komunitas TikTok di Indonesia."
]
tf = pd.DataFrame([Counter(doc.split()) for doc in documents],).fillna(0)
tf
```

```python
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```
